```python
import numpy as np
def NDD(x,y):
    #the function returns the interpolating polynomial by Newton's divided difference method
    n=len(x)
    coeff=np.zeros((n,n))
    coeff[0]=y
    for i in range(1,n):
        for j in range(i,n):
            coeff[i][j]=(coeff[i-1][j]-coeff[i-1][j-1])/(x[j]-x[j-i])
    co=np.zeros(n)
    for i in range(0,n):
        co[i]=coeff[i][i]
    return co
    
```

```python
x=[1.0,1.3,1.6,1.9,2.2]
y=[0.7651977,0.6200860,0.4554022,0.2818186,0.1103623]
```

```python
def prod(z,x,n):
    if n==0:
        return x[0]
    else:
        prod=1
        for i in range(n):
            prod=prod*(z-x[i])
    return prod
```

```python
def pol(z,x,co):
    #Z: flot x:vector co:vector
    sum=0
    for i in range(len(x)):
        sum=sum+co[i]*prod(z,x,i)
    return sum
    
```

```python
co=NDD(x,y)
print(co)
pol(1.5,x,co)
```
